```python
import numpy as np
import math as math
import matplotlib.pyplot as plt
from matplotlib import cm, ticker
```

```python
def rosbForce(q):
    # this function computes the negative gradient of the Rosenbrock
    # defined by the function zz =  (a-xx)**2 + b*(yy-xx**2)**2;

    global a,b #parameters
    
    # split off the x and y components of the vector q
    x = q[0]
    y = q[1]

    # the dimensions of f (it will be a 2d column vector)
    f = np.zeros(q.shape)

    # calculation of the x and y-components of the gradient
    dVx = -2*(a-x) - 4*b*(y-x**2)*x
    dVy =2*b*(y-x**2)

    
    # negate the gradient and return the result
    f[0]=-dVx 
    f[1]=-dVy
    return f
```

```python
q0 = np.array([0,0])
rosbForce(q0)
```

```python
# declare some global arrays used to hold 
# coefficients of model
global a,b
a = 1.
b=100

# creates matlab arrays xx, yy which contain 
# grid coordinates
xrange = np.arange(-8,8,0.05)
yrange = np.arange(-10,40,0.05)
xx,yy = np.meshgrid(xrange,yrange) 

# Compute the rosenbrock function at the 
# gridpoints
zz =  (a-xx)**2 + b*(yy-xx**2)**2

# clear the figure and draw a countour plot
plt.pcolor(xx,yy,zz,shading='auto', cmap='hsv')
plt.contour(xx,yy,zz, 40, cmap='jet')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Rosenbrock Banana')
plt.show()

```

```python
def Euler(q_n, h, vector_field):
# computes a step using Euler's method for the given vector field

    return q_n + h*vector_field(q_n)
```

```python
def compute_trajectory(q0, Nsteps, h,  step_function, vector_field):
# computes a trajectory of a differential equation, 
# starting from a given initial condition


# initialize trajectory
    q_traj = [q0]
    t_traj = [0]

# starting values
    q = np.copy(q0)
    t = 0

#loop
    for n in range(Nsteps):
        # calculate next step using provided stepping function
        q = step_function(q, h,  vector_field)
        t = t + h
        
        # update trajectory
        q_traj = q_traj + [q]
        t_traj = t_traj + [t]

    return np.array(q_traj),np.array(t_traj)
```

```python
# declare some global arrays used to hold 
# coefficients of model
global a,b
a = 1.
b=100

# creates matlab arrays xx, yy which contain 
# grid coordinates
xrange = np.arange(-4,4,0.05)
yrange = np.arange(-5,20,0.05)
xx,yy = np.meshgrid(xrange,yrange) 

# Compute the rosenbrock function at the 
# gridpoints
zz =  (a-xx)**2 + b*(yy-xx**2)**2

# clear the figure and draw a countour plot
plt.pcolor(xx,yy,zz,shading='auto', cmap='winter')
plt.contour(xx,yy,zz, 100, cmap='autumn')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Rosenbrock Banana')


q,t = compute_trajectory([2,5],100,0.01,Euler,rosbForce)
plt.plot(q[:,0], q[:,1], 'y')
plt.show()
```
